## Creating array of size 64*64 in range 0-255.

```python
## Creating array of size 64*64 in range 0-255.
import numpy as np
R = int(input("Enter the number of rows: "))
C = int(input("Enter the number of columns: "))    
array = np.random.randint(0,255, size=(R,C))

print(array)

print(type(array))
print(array.shape)
print(array.size)
print(array.shape[0])
```

## Spliting Array Column Wise

```python
"""
    This function will split the array into its column vectors and return a list 
    where each element of the list is the column vector of the array.
"""
array_split = []
## 64 vectors are created each containing 64 elements.
for i in range(0,64):
    vector = array[:,[i]]
    array_split.append(vector)
print(array_split)
```

## Mean

```python
"""
    This function finds the mean of each vector in our list and returns a list containing the mean of individual vectors in our list.
"""

## Creating an empty list.
mean_list = []
length_of_list   = len(array_split)
length_of_vector = len(array_split[0])
    
for i in range(0,length_of_list):
    SUM = 0
    vector = array_split[i]
        
    for j in range(0,length_of_vector):
        SUM = SUM + vector[j]
            
    mean = float(SUM/length_of_vector)
    mean_list.append(mean)

##Printing mean list :    
print(mean_list)
```

## Variance 

```python
#Calculating variance 
variance = []
length_of_list = len(array_split)
    
for i in range(0,length_of_list):
        
    vector1 = array_split[i]        
    vector1_mean = mean_list[i]       
    vector1_minus_mean = vector1 - vector1_mean       
    vector1_minus_mean_sqaure = vector1_minus_mean**2       
    SUM = 0
    
    for j in range(0,vector1_minus_mean_sqaure.shape[0]):
        SUM = SUM + vector1_minus_mean_sqaure[j]       
    final_Variance = float(SUM/vector1_minus_mean_sqaure.shape[0])
    variance.append(final_Variance)

## Prinitng variance   
print(variance)
```

## Covariance

```python
## Calculating covariance of vectors :

v1 = int(input("Enter vector choice: "))
v2 = int(input("Enter vector choice: "))

vector2 = array_split[v1]   
vector3 = array_split[v2]
    
vector2_mean = mean_list[v1]
vector3_mean = mean_list[v2]
    
vector2_minus_mean = vector2 - vector2_mean
vector3_minus_mean = vector3 - vector3_mean
    
vector4 = vector2_minus_mean * vector3_minus_mean    
length = vector4.shape[0]
SUM = 0
for i in range(0,length):
    SUM = SUM + vector4[i]
Covariance = float(SUM/length)
        
print(Covariance)
```

## Multiplication

```python
## Multiplication of two vector
"""
    It will take input as any two vectors and print their product.
"""
v1 = int(input("Enter vector choice: "))
v2 = int(input("Enter vector choice: "))
    
vector5 = array_split[v1] 
vector6 = array_split[v2]
            
temp = vector5 * vector6
SUM = 0
for i in range(0,temp.shape[0]):
    SUM = SUM + temp[i]
    print(SUM)
```
